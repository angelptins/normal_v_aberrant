## # Functions

```python
def get_xyz(df, kp, dims=3):
    
    x = df[f'{kp}_x'].values
    y = df[f'{kp}_y'].values
    
    if dims==2:
        return x, y
    elif dims==3:
        z = df[f'{kp}_z'].values
        return x, y, z
```

```python
def get_mags(df, kp, dims=3):
    return np.linalg.norm( 
        df.filter(like=kp).iloc[:,:dims].values,
        axis=1
    )
```

```python
def get_velocity(x, y, z=None):
    
    vx    = np.gradient(x)
    vy    = np.gradient(y)
    speed = np.sqrt(vx**2 + vy**2)
    
    if z is None:
        return vx, vy, speed
    else:
        vz    = np.gradient(z)
        speed = np.sqrt(vx**2 + vy**2 + vz**2)
        return vx, vy, vz, speed
```

```python
def get_acceleration(vx, vy, speed, vz=None):
    
    ax = np.abs( np.gradient(vx) )
    ay = np.abs( np.gradient(vy) )
    a  = np.abs( np.gradient(speed) )
    
    if vz is None:
        return ax, ay, a
    else:
        az = np.gradient(vz)
        return ax, ay, az, a
```

```python
def get_jerk(ax, ay, a, az=None):
    
    jx = np.abs( np.gradient(ax) )
    jy = np.abs( np.gradient(ay) )
    j  = np.abs( np.gradient(a) )
    
    if az is None:
        return jx, jy, j
    else:
        jz = np.gradient(jz)
        return jx, jy, jz, j
```

```python
def get_distances_between_keypoints( df, kp1, kp2, dims=3 ):
    
    x1, y1, z1 = get_xyz( df, kp1 )
    x2, y2, z2 = get_xyz( df, kp2 )
    
    if dims==2:
        distances = np.sqrt( (x2 - x1)**2 + (y2 - y1)**2  )
    elif dims==3:
        distances = np.sqrt( (x2 - x1)**2 + (y2 - y1)**2 + (z2 - z1)**2 )
    
    return distances
```

```python
def get_angles_between_keypoints(df, kp1, kp2, kp3 ):
    
    A = df[[f'{kp1}_x', f'{kp1}_y']].values
    B = df[[f'{kp2}_x', f'{kp2}_y']].values
    C = df[[f'{kp3}_x', f'{kp3}_y']].values

    BA = A - B
    BC = C - B

    dot = np.sum(BA * BC, axis=1)
    magBA = np.linalg.norm(BA, axis=1)
    magBC = np.linalg.norm(BC, axis=1)

    cos_angle = dot / (magBA * magBC)
    cos_angle = np.clip(cos_angle, -1.0, 1.0)  # numerical safety

    angles = np.degrees(np.arccos(cos_angle))

    return angles
```

```python
def get_torso_areas(df):
    
    x = df[[
        'left_shoulder_x', 'right_shoulder_x',
        'right_hip_x', 'left_hip_x',
    ]].values

    y = df[[
        'left_shoulder_y', 'right_shoulder_y',
        'right_hip_y', 'left_hip_y',
    ]].values

    areas = 0.5 * np.abs(
        np.sum(x * np.roll(y, -1, axis=1), axis=1) -
        np.sum(y * np.roll(x, -1, axis=1), axis=1)
    )

    return areas
```

```python
def transform_df( 
        df,
        kps=[
            'left_wrist','right_wrist',
            'left_ankle','right_ankle',
            # 'left_elbow','right_elbow',
            # 'left_knee','right_knee',
            # 'center_shoulder',
        ], 
        kp_pairs=[
            ### Contralateral ###
            ('left_wrist','right_wrist'),
            ('left_ankle','right_ankle'),
            
            ### Ipsilateral ###
            # ('nose','left_wrist'),
            # ('nose','right_wrist'),
            # ('nose','left_shoulder'),
            # ('nose','right_shoulder'),
            ## Upper Mobility
            ('left_shoulder','left_wrist'),
            ('right_shoulder','right_wrist'),
            ## Lower Mobility
            ('left_hip','left_ankle'),
            ('right_hip','right_ankle'),

            ## Miscellaenous
            ('left_wrist','left_ankle'),
            ('right_wrist','right_ankle'),
            # ('left_wrist', 'center_shoulder'),
            # ('right_wrist', 'center_shoulder'),
            # ('left_ankle', 'center_hip'),
            # ('right_ankle', 'center_hip'),
        ],
        kp_triples_for_external_angles=[
            ('center_shoulder','left_shoulder','left_elbow'),
            ('center_shoulder','right_shoulder','right_elbow'),
            ('center_hip','left_hip','left_knee'),
            ('center_hip','right_hip','right_knee'),
        ],
        # kp_triples_for_internal_angles = [
        #     ('right_shoulder','left_shoulder','left_hip'), # UL
        #     ('left_shoulder','right_shoulder','right_hip'),  # UR
        #     ('right_hip','left_hip','left_shoulder'), # BL
        #     ('left_hip','right_hip','right_shoulder'),  # BR
        # ],
    ):

    values = []
    colnames = []
    
    for kp in kps:

        x, y, z = get_xyz( df, kp )

        mags_2d = get_mags( df, kp, dims=2)
        # mags_3d = get_mags( df, kp, dims=3)
        
        vx, vy, speed = get_velocity( x, y )
        ax, ay, accel = get_acceleration( vx, vy, speed )
        _, _,   jerk  = get_jerk( ax, ay, accel )

        values.append( x )
        values.append( y )
        # values.append( z )
        
        values.append(mags_2d)
        # values.append(mags_3d)
        
        values.append(speed)
        values.append(accel)
        values.append(jerk)

        colnames.append( f'{kp}_x' )
        colnames.append( f'{kp}_y' )
        # colnames.append( f'{kp}_z' )
        
        colnames.append( f'{kp}_mag_2d' )
        # colnames.append( f'{kp}_mag_3d' )
        
        colnames.append( f'{kp}_speed' )
        colnames.append( f'{kp}_accel' )
        colnames.append( f'{kp}_jerk' )

    # Distances b/t keypoints
    for kp1, kp2 in kp_pairs:
        distances = get_distances_between_keypoints( df, kp1, kp2 )
        values.append( distances )
        colnames.append( f'{kp1}_{kp2}_dist' )

    # Angles b/t Joint Keypoints ("external")
    for kp1, kp2, kp3 in kp_triples_for_external_angles:
        angles = get_angles_between_keypoints( df, kp1, kp2, kp3 )
        values.append( angles )
        colnames.append( f'{kp2}_external_angle' )

    # Angles b/t Torso Keypoints ("internal")
    # for kp1, kp2, kp3 in kp_triples_for_internal_angles:
    #     angles = get_angles_between_keypoints( df, kp1, kp2, kp3 )
    #     values.append( angles )
    #     colnames.append( f'{kp2}_internal_angle' )

    # Area of Torso Box for Torso Flexion
    areas = get_torso_areas( df )
    values.append( areas )
    colnames.append( 'torso_area' )
    
    df_transformed = pd.DataFrame( values ).T
    df_transformed.columns = colnames
    return df_transformed
```

## # Metrics

```python
class Specificity(Metric):
    def __init__(self, pos_label=1):
        self.pos_label = pos_label

    def reset(self):
        self.tn = 0
        self.fp = 0

    def accumulate(self, learn):
        preds = learn.pred
        targs = learn.y

        # Convert probabilities → class predictions if needed
        if preds.ndim > 1:
            preds = preds.argmax(dim=1)

        neg = targs != self.pos_label
        pred_neg = preds != self.pos_label

        self.tn += (pred_neg & neg).sum().item()
        self.fp += (~pred_neg & neg).sum().item()

    @property
    def value(self):
        denom = self.tn + self.fp
        return self.tn / denom if denom > 0 else 0.0

    @property
    def name(self):
        return "specificity"
```

## # Begin Modeling

```python
from fastai.callback.tracker import SaveModelCallback
```

```python
def run_experiment(
        X_np,
        y_labels_np,
        idxs_SamplesBalanced,
        skip=3,
        bs=8,
        n_epochs=200,
        random_seed=0,
    ):

    # -----------------------
    # 1. Reproducibility
    # -----------------------
    set_seed(random_seed, reproducible=True)

    X_np_balanced        = X_np[idxs_SamplesBalanced, :, ::skip]
    y_labels_np_balanced = y_labels_np[idxs_SamplesBalanced]

    # -----------------------
    # 2. Splits
    # -----------------------
    splits = get_splits(
        y_labels_np_balanced,
        valid_size=0.3,
        stratify=True,
        random_state=random_seed
    )

    # -----------------------
    # 3. Datasets
    # -----------------------
    dsets = TSDatasets(
        X_np_balanced,
        y_labels_np_balanced,
        tfms=[None, [Categorize()]],
        splits=splits,
        inplace=True
    )

    # -----------------------
    # 4. DataLoaders (fully deterministic)
    # -----------------------
    dls = TSDataLoaders.from_dsets(
        dsets.train, dsets.valid,
        bs=bs,
        batch_tfms=[TSStandardize(by_sample=False, by_var=False)],
        num_workers=0,
        shuffle_train=True,
        shuffle_valid=False,
    )

    # -----------------------
    # 5. Model + Learner
    # -----------------------
    model = InceptionTime(dls.vars, dls.c)

    # IMPORTANT: stable metric names
    metrics = [
        accuracy,
        RocAucBinary(),
        Precision(),
        Specificity(),
        Recall(),
        F1Score()
    ]

    # -----------------------
    # 6. Safe checkpointing (NO manual mv/rename)
    # -----------------------
    save_best_f1 = SaveModelCallback(
        monitor='f1_score',  
        fname=f'best_f1_seed{random_seed:03d}',
        comp=np.greater
    )
    
    save_best_auc = SaveModelCallback(
        monitor='roc_auc_score', 
        fname=f'best_auc_seed{random_seed:03d}',
        comp=np.greater
    )

    learn = Learner(
        dls,
        model,
        metrics=metrics,
    )

    # -----------------------
    # 7. Train (deterministic LR instead of lr_find)
    # -----------------------
    lr_max = 1e-3

    with learn.no_bar(), learn.no_logging():
        learn.fit_one_cycle(n_epochs, lr_max=lr_max, cbs=[save_best_f1, save_best_auc],
    )

    # -----------------------
    # 8. Save final model safely
    # -----------------------
    last_model_name = f'last_model_seed{random_seed:03d}'
    learn.save(last_model_name)

    # -----------------------
    # 9. Training metrics export
    # -----------------------
    os.makedirs('./df_metrics/', exist_ok=True)

    df_metrics = pd.DataFrame(
        learn.recorder.values,
        columns=learn.recorder.metric_names[1:-1]
    )

    df_metrics.to_csv(
        f'./df_metrics/df_train_metrics_seed{random_seed:03d}.csv',
        index=False
    )

    # -----------------------
    # 10. VALIDATION (BEST MODEL - F1)
    # -----------------------
    
    learn.load(f'best_f1_seed{random_seed:03d}')
    metric_names = ['valid_loss'] + [m.name for m in learn.metrics]

    val_results_f1 = learn.validate()
    df_val = pd.DataFrame([val_results_f1], columns=metric_names)
    df_val.to_csv(
        f'./df_metrics/df_val_metrics_best_f1_seed{random_seed:03d}.csv',
        index=False
    )

    learn.load(f'best_auc_seed{random_seed:03d}')
    metric_names = ['valid_loss'] + [m.name for m in learn.metrics]

    val_results_auc = learn.validate()
    df_val = pd.DataFrame([val_results_auc], columns=metric_names)
    df_val.to_csv(
        f'./df_metrics/df_val_metrics_best_auc_seed{random_seed:03d}.csv',
        index=False
    )

    return df_metrics
```

```python
num_trials = 20
```

```python
for random_seed in range( num_trials ):
    print(f'{random_seed}/{num_trials}')
    _ = run_experiment(
        X_np,
        y_labels_np,
        idxs_SamplesBalanced,
        skip=3,
        random_seed=random_seed,
    )
```
